```python
# Qubits / Block Sphere / Basis States 
import qiskit 
import numpy as np
qiskit.__qiskit_version__
```

```python
# Qubits are the basic unit of computaiton in QuantComp 
# Used to work with QuantComp in diff levels of abstraction 
# 
# |0> = ( 1 )
#  .... ( 0 )
#
# |1> = ( 0 )
# ..... ( 1 )
from qiskit import *
# Representing Qubits 
# Visualise with Bloch Sphere 
from qiskit.visualization import plot_bloch_vector
plot_bloch_vector( [ 0,0,1 ], title='aSphere')
```

```python
# Spin + / - 
# + : |+> = ( 1/(sqrt)2 ) =    1
# ...     = ( 1/(sqrt)2 )   -------( |0>+|1> )
# ...                       (sqr)2 
#
# + : |-> = ( 1/(sqrt)2 ) =     1
# ...     = ( -1/(sqrt)2 )   -------( |0>-|1> )
# ...                         (sqr)2 

plot_bloch_vector( [ 1,0,0 ] )
```

```python
# Basis State 

ketZero = np.array([ [ 1 ],[ 0 ] ])
ketOne = np.array([ [ 0 ],[ 1 ] ])

np.kron( ketOne, ketZero )
```
